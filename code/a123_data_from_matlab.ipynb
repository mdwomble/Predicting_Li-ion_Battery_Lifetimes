# Prepping a123 Data from Matlab Files

### Import Modules

```python
import h5py
import pandas as pd
import numpy as np
import pickle
```

### Load Matlab Files

```python
# set path to a123 matlab files
file_path=r'C:\Users\dmmkm\Documents\Projects\battery_management_systems\a123_matlab_files'
timeseries_path=r'C:\Users\dmmkm\Documents\Projects\battery_management_systems\timeseries_data\a123'
cycle_path=r'C:\Users\dmmkm\Documents\Projects\battery_management_systems\cycling_data\a123'
```

```python
# Batch 1
matFilename = file_path + '/2017-05-12_batchdata_updated_struct_errorcorrect.mat'
f1 = h5py.File(matFilename)

# Batch 2
matFilename = file_path + '/2017-06-30_batchdata_updated_struct_errorcorrect.mat'
f2 = h5py.File(matFilename)

# Batch 3
matFilename = file_path + '/2018-04-12_batchdata_updated_struct_errorcorrect.mat'
f3 = h5py.File(matFilename)
```

### Batch 1
code written by rdbratz and pulled from github

```python
# Get batch data from h5py file
batch = f1['batch']

# Put Timeseries data into dictionary
num_cells = batch['summary'].shape[0]
bat_dict1 = {}
for i in range(num_cells):
    print('Cell:', i)
    cl = f1[batch['cycle_life'][i,0]][()]
    policy = f1[batch['policy_readable'][i,0]][()].tobytes()[::2].decode()
    summary_IR = np.hstack(f1[batch['summary'][i,0]]['IR'][0,:].tolist())
    summary_QC = np.hstack(f1[batch['summary'][i,0]]['QCharge'][0,:].tolist())
    summary_QD = np.hstack(f1[batch['summary'][i,0]]['QDischarge'][0,:].tolist())
    summary_TA = np.hstack(f1[batch['summary'][i,0]]['Tavg'][0,:].tolist())
    summary_TM = np.hstack(f1[batch['summary'][i,0]]['Tmin'][0,:].tolist())
    summary_TX = np.hstack(f1[batch['summary'][i,0]]['Tmax'][0,:].tolist())
    summary_CT = np.hstack(f1[batch['summary'][i,0]]['chargetime'][0,:].tolist())
    summary_CY = np.hstack(f1[batch['summary'][i,0]]['cycle'][0,:].tolist())
    summary = {'IR': summary_IR, 'QC': summary_QC, 'QD': summary_QD, 'Tavg':
                summary_TA, 'Tmin': summary_TM, 'Tmax': summary_TX, 'chargetime': summary_CT,
                'cycle': summary_CY}
    cycles = f1[batch['cycles'][i,0]]
    cycle_dict = {}
    for j in range(cycles['I'].shape[0]):
        I = np.hstack((f1[cycles['I'][j,0]][()]))
        Qc = np.hstack((f1[cycles['Qc'][j,0]][()]))
        Qd = np.hstack((f1[cycles['Qd'][j,0]][()]))
        T = np.hstack((f1[cycles['T'][j,0]][()]))
        V = np.hstack((f1[cycles['V'][j,0]][()]))
        t = np.hstack((f1[cycles['t'][j,0]][()]))
        cd = {'I': I, 'Qc': Qc, 'Qd': Qd, 'T': T, 'V':V, 't':t}
        cycle_dict[str(j)] = cd
        
    cell_dict = {'cycle_life': cl, 'charge_policy':policy, 'summary': summary, 'cycles': cycle_dict}
    key = 'b1c' + str(i)
    bat_dict1[key] = cell_dict
```

```python
#remove batteries that do not reach 80% capacity
del bat_dict1['b1c8']
del bat_dict1['b1c10']
del bat_dict1['b1c12']
del bat_dict1['b1c13']
del bat_dict1['b1c22']
```

```python
# There are four cells from batch1 that carried into batch2, we'll remove the data from batch2
# and put it with the correct cell from batch1
bat_dict2_keys = ['b2c7', 'b2c8', 'b2c9', 'b2c15', 'b2c16']
bat_dict1_keys = ['b1c0', 'b1c1', 'b1c2', 'b1c3', 'b1c4']
add_len = [662, 981, 1060, 208, 482]

for i, bk in enumerate(bat_dict1_keys):
    bat_dict1[bk]['cycle_life'] = bat_dict1[bk]['cycle_life'] + add_len[i]
    for j in bat_dict1[bk]['summary'].keys():
        if j == 'cycle':
            bat_dict1[bk]['summary'][j] = np.hstack((bat_dict1[bk]['summary'][j], bat_dict2[bat_dict2_keys[i]]['summary'][j] + len(bat_dict1[bk]['summary'][j])))
        else:
            bat_dict1[bk]['summary'][j] = np.hstack((bat_dict1[bk]['summary'][j], bat_dict2[bat_dict2_keys[i]]['summary'][j]))
    last_cycle = len(bat_dict1[bk]['cycles'].keys())
    for j, jk in enumerate(bat_dict2[bat_dict2_keys[i]]['cycles'].keys()):
        bat_dict1[bk]['cycles'][str(last_cycle + j)] = bat_dict2[bat_dict2_keys[i]]['cycles'][jk]
        
del bat_dict2['b2c7']
del bat_dict2['b2c8']
del bat_dict2['b2c9']
del bat_dict2['b2c15']
del bat_dict2['b2c16']
```

### Create Dataframes

```python
# combine batch dictionaries
bat_dict = {**bat_dict1, **bat_dict2, **bat_dict3}
```

##### Timeseries Data

```python
# Cycle through battery dictionary and extract timeseries data
for cell in bat_dict.keys():
    df=pd.DataFrame()
    for cycle in bat_dict[cell]['cycles']:
        print(cell+': '+cycle)
        dt=[0]
        
        if int(cycle) > 0:
            df_new=pd.DataFrame.from_dict(bat_dict[cell]['cycles'][cycle])
            df_new['Cycle_Index']=int(cycle)
            dt=dt+list(np.diff(df_new['t'].to_numpy()))
            df_new['dt']=dt
            df=pd.concat([df, df_new])

    df=df[df['Cycle_Index']>0]
    df.rename(columns={'I':'Current (A)', 'Qc':'Charge_Capacity (Ah)', 'Qd':'Discharge_Capacity (Ah)', 'T':'Cell_Temperature (C)', 'V':'Voltage (V)', 't':'Test_Time (s)'}, inplace=True)
    df.reset_index(drop=True)
    
    df.to_csv(timeseries_path+'\\a123_'+cell+'.csv', index=False)
```

##### Cycling Data

```python
# Cycle through battery dictionary and extract cycling data
for cell in bat_dict.keys():
    df=pd.DataFrame.from_dict(bat_dict[cell]['summary'])
    df.rename(columns={'Qc':'Charge_Capacity (Ah)', 'Qd':'Discharge_Capacity (Ah)'}, inplace=True)
    
    df.to_csv(cycle_path+'\\'+cell+'.csv', index=False)
```
